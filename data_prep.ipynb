```python
import ast
import glob
import pandas as pd
```

```python
def load_dictionary_from_file(file_path):
    try:
        with open(file_path, 'r') as file:
            dictionary_str = file.read()
            dictionary = ast.literal_eval(dictionary_str)
            if isinstance(dictionary, dict):
                return dictionary
            else:
                print("Error: The file does not contain a valid dictionary.")
                return None
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

# Example usage:
file_paths = glob.glob('*.txt')
loaded_dictionaries = [load_dictionary_from_file(fp) for fp in file_paths]
```

```python
frames = [pd.DataFrame.from_dict(ld).transpose() for ld in loaded_dictionaries]
```

```python
big_frame = pd.concat(frames, axis = 0)
```

```python
big_frame.head()
```

```python
for c in big_frame.columns :
    if big_frame[c].str.contains('\n').any() :
        try :
            big_frame[c] = big_frame[c].apply(lambda string : string.split('\n')[1])
        except IndexError :
            pass
```

```python
big_frame.head()
```

```python
for c in big_frame.columns :
    big_frame[c] = big_frame[c].astype(str)
```

```python
df_no_duplicates = big_frame.drop_duplicates()
```
